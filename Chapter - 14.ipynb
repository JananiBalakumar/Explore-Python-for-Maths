```python
import sympy as sp
```

```python
dir(sp)
```

```python
# Define a symbolic variable
x = sp.Symbol('x')

# Define a symbolic expression
expr = x**2 + 2*x + 1

# Perform algebraic manipulation
simplified_expr = sp.simplify(expr)

# Solve an equation
solution = sp.solve(expr, x)

print(simplified_expr)  
print(solution)        
```

```python
# Define symbolic variables
x, y = sp.symbols('x y')

# Define an algebraic expression
expr = x**2 + 2*x + 1

# Find the derivative of the expression with respect to x
derivative_expr = sp.diff(expr, x)

# Integrate the expression with respect to x
integral_expr = sp.integrate(expr, x)

# Solve a simple equation x^2+3x+2 = 0
equation = sp.Eq(x**2 + 3*x + 2,0)
solution = sp.solve(equation, x)

# Print the results
print("Expression:", expr)
print("Derivative:", derivative_expr)
print("Integral:", integral_expr)
print("Solution of the Equation:", solution)
```

```python
import numpy as np
from scipy.optimize import minimize

# Define the objective function to minimize
def objective(x):
    return x**2 + 5 * np.sin(x)

# Initial guess for the minimum
x0 = 2.0

# Perform the optimization
result = minimize(objective, x0)

# Extract the optimized value
min_x = result.x[0]
min_y = result.fun

print("Optimized Minimum Value:")
print("x =", min_x)
print("f(x) =", min_y)
```

```python
import numpy as np
from scipy.stats import norm, ttest_ind

# Sample data for two groups
group1 = np.array([82, 87, 85, 90, 78])
group2 = np.array([75, 88, 92, 79, 85])

# Calculate mean and standard deviation
mean_group1 = np.mean(group1)
mean_group2 = np.mean(group2)
# Use ddof=1 for sample standard deviation
std_group1 = np.std(group1, ddof=1)  
std_group2 = np.std(group2, ddof=1)

print("Group 1:")
print("Mean:", mean_group1)
print("Standard Deviation:", std_group1)

print("\nGroup 2:")
print("Mean:", mean_group2)
print("Standard Deviation:", std_group2)

# Perform a two-sample t-test
t_stat, p_value = ttest_ind(group1, group2)

print("\nTwo-sample t-test:")
print("T-statistic:", t_stat)
print("P-value:", p_value)
```

```python
import numpy as np
import statsmodels.api as sm

# Sample data
x = np.array([1, 2, 3, 4, 5])
y = np.array([2, 4, 5, 4, 5])

# Add constant term to the independent variable (required for OLS)
x = sm.add_constant(x)

# Fit linear regression model
model = sm.OLS(y, x).fit()

# Print model summary
print(model.summary())
```

```python
import numpy as np
import statsmodels.api as sm

# Sample data
X1 = np.array([1, 2, 3, 4, 5])
X2 = np.array([2, 4, 5, 7, 6])
Y = np.array([5, 7, 8, 10, 11])

# Create a design matrix by adding a constant term to the independent variables
X = np.column_stack((X1, X2))
X = sm.add_constant(X)

# Fit multiple linear regression model
model = sm.OLS(Y, X).fit()

# Print model summary
print(model.summary())
```

```python
import pandas as pd

# Create a DataFrame
data = {
    'A': [10, 20, 30, 40, 50],
    'B': [5, 15, 25, 35, 45],
    'C': [3, 6, 9, 12, 15]
}

df = pd.DataFrame(data)

# Perform basic arithmetic operations
df['Sum'] = df['A'] + df['B']
df['Product'] = df['A'] * df['C']
df['Mean'] = df.mean(axis=1)

# Calculate summary statistics
mean_col = df.mean()
sum_row = df.sum()

print("DataFrame:")
print(df)

print("\nMean for each column:")
print(mean_col)

print("\nSum for each row:")
print(sum_row)
```
