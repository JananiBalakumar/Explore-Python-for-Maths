```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

# Data Creation
df = pd.DataFrame([['1', 9, 7, 10], 
                   ['2', 9, 9, 8], 
                   ['3', 10, 10, 8]],
                  columns=['Team', 'Round 1', 'Round 2', 'Round 3'])

# Convert data to percentages
df_percent = df.set_index('Team').div(df.set_index('Team').sum(axis=1), axis=0) * 100

# Plot stacked percentage bar chart
ax = df_percent.plot(kind='bar', stacked=True,
                     title='Stacked Percentage Bar Chart', width=0.3)

# Add labels to the bars
for p in ax.patches:
    width, height = p.get_width(), p.get_height()
    x, y = p.get_xy() 
    ax.annotate(f'{height:.1f}%', (x + width / 2, y + height / 2), 
                ha='center', va='center')

plt.ylabel('Percentage')
plt.show()
```

```python
# importing packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# load dataset
data = {'Category':['A','B','C'],
        'val1': [10,20,30],
        'val2': [15,25,35],
        'val3': [9,18,27]
        
}
df = pd.DataFrame(data)
# view dataset
print(df)

# plot a Stacked Bar Chart using matplotlib
df.plot(x = 'Category',kind = 'barh',stacked = True,
title = 'Percentage Stacked Bar Graph',
mark_right = True)

df_total = df["val1"] + df["val2"] + df["val3"]
df_rel = df[df.columns[1:]].div(df_total, 0)*100

for n in df_rel:
    for i, (cs, ab, pc) in enumerate(zip(df.iloc[:, 1:].cumsum(1)[n],
               df[n], df_rel[n])):
    plt.text(cs - ab / 2, i, str(np.round(pc, 1)) + '%',
          va = 'center', ha = 'center')
```

#### 15.5.2 Cumulative Histogram

```python
# importing 
import numpy as np
from scipy import stats
 
# list of values
x = [10, 50, 60, 10, 10, 12, 69]
res = stats.cumfreq(x, numbins=4,
                    defaultreallimits=(1.5, 5))
# generating random values
rng = np.random.RandomState(seed=12345)
# normalizing
samples = stats.norm.rvs(size=1000,
                         random_state=rng)
res = stats.cumfreq(samples,
                    numbins=25)
x = res.lowerlimit + np.linspace(0, res.binsize*res.cumcount.size,
                                 res.cumcount.size)
 
fig = plt.figure(figsize=(10, 4))
ax1 = fig.add_subplot(1, 2, 1)
ax2 = fig.add_subplot(1, 2, 2)
ax1.hist(samples, bins=20, color="purple")
ax1.set_title('Histogram')
ax2.bar(x, x, width=2, color="red")
ax2.set_title('Cumulative Histogram')
ax2.set_xlim([x.min(), x.max()])
plt.show()
```

#### 15.7.2 Customizing the Pie chart

```python
# Import libraries
import numpy as np
import matplotlib.pyplot as plt

# Creating dataset
subjects = ['Python','Maths', 
            'Statistics', 'Algebra']
data = [39,28,36, 30]
# Creating explode data
explode = (0.1, 0.0, 0.2, 0.3)
# Creating color parameters
colors = ( "purple", "yellow", "brown","pink")
# Wedge properties
wp = { 'linewidth' : 1, 'edgecolor' : "cyan" }
# Creating autocpt arguments
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%\n({:d} g)".format(pct, absolute)
# Creating plot
fig, ax = plt.subplots(figsize =(10, 7))
wedges, texts, autotexts = ax.pie(data,
                                  autopct = lambda pct: func(pct, data),
                                  explode = explode,
                                  labels = subjects,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="magenta"))

# Adding legend
ax.legend(wedges, subjects,
          title ="Subjects",
          loc ="center left",
          bbox_to_anchor =(1, 0, 0.5, 1))
plt.setp(autotexts, size = 8, weight ="bold")
ax.set_title("Customizing pie chart")
plt.show()
```

#### 15.8.3 3D Scatter Plot

```python
# Import libraries
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
 
# Creating dataset
z = 4 * np.tan(np.random.randint(10, size =(100))) 
x = 4 * np.cos(z) + np.random.normal(size = 100)
y = 4 * np.sin(z) + 4 * np.random.normal(size = 100)
# Creating figure
fig = plt.figure(figsize = (16, 9))
ax = plt.axes(projection ="3d")
# Add x, y gridlines
ax.grid(b = True, color ='grey',
        linestyle ='-.', linewidth = 0.3,
        alpha = 0.2) 
# Creating color map
my_cmap = plt.get_cmap('hsv')
 
# Creating plot
sctt = ax.scatter3D(x, y, z,
                    alpha = 0.8,
                    c = (x + y + z),
                    cmap = my_cmap,
                    marker ='*')
plt.title("simple 3D scatter plot")
ax.set_xlabel('X-axis', fontweight ='bold')
ax.set_ylabel('Y-axis', fontweight ='bold')
ax.set_zlabel('Z-axis', fontweight ='bold')
fig.colorbar(sctt, ax = ax, shrink = 0.5, aspect = 5)
# show plot
plt.show()
```

#### 15.8.6 3D Contour Plot

```python
from mpl_toolkits import mplot3d
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
import math
  
x = [i for i in range(0, 200, 100)]
y = [i for i in range(0, 200, 100)]
  
X, Y = np.meshgrid(x, y)
Z = []
for i in x:
    t = []
    for j in y:
        t.append(math.cos(math.sqrt(i*2+j*2)))
    Z.append(t)
  
fig = plt.figure()
ax = plt.axes(projection='3d')
ax.contour3D(X, Y, Z, 50, cmap=cm.cool)
ax.set_xlabel('a')
ax.set_ylabel('b')
ax.set_zlabel('c')
ax.set_title('3D contour for cosine')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the function to plot (3D Gaussian-like function)
def f(x, y):
    return np.exp(-(x**2 + y**2)) * np.cos(5 * np.sqrt(x**2 + y**2))

# Create a meshgrid of points
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Create a contour plot
contour = ax.contourf(X, Y, Z, cmap='viridis')

# Add colorbar
cbar = plt.colorbar(contour)

# Set labels
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

# Set title
ax.set_title('3D Contour Plot Example')

# Show the plot
plt.show()
```

#### 15.9.3 Creating multiple plots with different sizes

```python
# importing required libraries
import matplotlib.pyplot as plt
from matplotlib import gridspec
import numpy as np
 
# create a figure
fig = plt.figure()
# set height of each subplot as 8
fig.set_figheight(8)
# set width of each subplot as 8
fig.set_figwidth(8)
 
# create grid for different subplots
spec = gridspec.GridSpec(ncols=2, nrows=2,
                         width_ratios=[2, 1], wspace=0.5,
                         hspace=0.5, height_ratios=[1, 2])
 
# initializing x,y axis value
x = np.arange(0, 10, 0.1)
y = np.sin(x)
 
ax0 = fig.add_subplot(spec[0])
ax0.plot(x, y, color = 'red')
ax1 = fig.add_subplot(spec[1])
ax1.plot(x, y, color = 'aqua')
ax2 = fig.add_subplot(spec[2])
ax2.plot(x, y, color = 'gold')
ax3 = fig.add_subplot(spec[3])
ax3.plot(x, y, color = 'indigo')
# display the plots
plt.show()
```
