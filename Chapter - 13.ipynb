# Probability Distribution

### 13.2 Normal Distribution

```python
from scipy.stats import norm
import numpy as np
import matplotlib.pyplot as plt
```

```python
norm_dist = norm.rvs(size = 100)
norm_dist
```

```python
import seaborn as sns
```

```python
sns.displot(norm_dist, kde = True)
```

### 13.3 Binomial Distribution

```python
n,p = 5,0.5
bin_dist = np.random.binomial(n,p,100)
bin_dist
```

```python
sns.histplot(bin_dist)
plt.show()
```

### 13.4 Poisson Distribution

```python
poi_dist = np.random.poisson(lam = 7, size = 100)
poi_dist
```

```python
sns.histplot(poi_dist, kde = False)
plt.show()
```

### 13.5 Exponential Distribution

```python
from scipy.stats import expon
ex_dist = expon.rvs(size = 50, scale = 20)
ex_dist
```

```python
sns.histplot(ex_dist)
```

### 13.6 Geometric Distribution

### Method 1

```python
size = 100
s = np.random.geometric(0.5, size)
s
```

```python
sns.histplot(s)
plt.show()
```

### Method 2

```python
from scipy.stats import geom
geo = geom.rvs(0.5, 0,100)
geo
```

```python
sns.histplot(geo)
plt.show()
```
